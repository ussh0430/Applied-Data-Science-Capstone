Import required modules

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Request and get data from website and pass data to BeautifulSoup 

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
html = requests.get(url).text
soup = BeautifulSoup(html, 'lxml')
```

Scrape website data for Postal Codes, Borough, and Neighbourhood values (marked with 'th' and 'td')

```python
for table in soup.find_all('table', class_= 'wikitable sortable') :
    col_name = []
    for col in table.find_all('th'):
        col_name.append(col.text.rstrip('\n'))
    
    codes = []
    for entry in table.find_all('td'):
        codes.append(entry.text.rstrip('\n'))
    
    data = [ codes [i:i + 3] for i in range(0, len(codes), 3) ]
    
df = pd.DataFrame(data, columns=col_name)
df
```

Drop Boroughs that are 'Not assigned' and copy Borough value when Neighbourhood has 'Not assigned' value

```python
df1=df[df.Borough != 'Not assigned']
df1.loc[df.Neighbourhood == 'Not assigned', 'Neighbourhood'] = df.Borough
df1
```

Groupby Postal Codes and combine/join Neighbourhood value 

```python
df1_group= df1.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
df2= df1_group.apply(pd.Series)
#pd.set_option('display.max_rows', 1000)
df2
```
